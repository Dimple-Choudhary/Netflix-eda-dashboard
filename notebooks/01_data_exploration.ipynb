```python
print(f"Before removing duplicates: {df.shape}")
df.drop_duplicates(inplace=True)
print(f"After removing duplicates: {df.shape}")
```

```python
df.isnull().sum()
```

```python
for col in ['rating', 'duration', 'director', 'cast', 'country']:
    df[col] = df[col].fillna("unknown")
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])

```

```python
df.isnull().sum()
```

```python
print("Initial dtype:", df['date_added'].dtype)
print("Sample values:")
display(df['date_added'].head(10))
```

```python
df['date_added_raw'] = df['date_added']
df['date_added'] = df['date_added'].astype(str).str.strip()
df.loc[df['date_added'].isin(['', 'nan', 'None', 'NaN', 'NoneType']), 'date_added'] = np.nan
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
num_failed = df['date_added_parsed'].isna().sum()
print(f"Parsed OK: {df.shape[0] - num_failed}, Failed/NaT: {num_failed}")
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

print("Final dtype:", df['date_added'].dtype)
print("Number of NaT in date_added:", df['date_added'].isna().sum())
df[['title', 'date_added', 'year_added', 'month_added']]
```

```python
text_cols = ['title', 'director', 'cast', 'country', 'rating', 'listed_in']

for col in text_cols:
  df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df['duration_int'] = df['duration'].str.extract('(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract('([A-Za-z]+)').astype(str)

```

```python
df.info()
df.sample(5)
```

```python
from datetime import datetime

df['content_age'] = datetime.now().year - df['release_year']
df['is_modern'] = df['release_year'] >= 2015
df['num_genres'] = df['listed_in'].apply(lambda x: len(x.split(',')))
df['has_known_director'] = df['director'] != 'Unknown'
```

```python
df.to_csv("../data/netflix_cleaned.csv", index=False)
print("âœ… Cleaned dataset saved successfully!")
```
